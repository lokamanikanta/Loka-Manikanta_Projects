```python
q1 = l2[3]
q3 = l2[5]
print(q1)
print(q3)
IQR =q3-q1
UF1 = q3+(1.5)*IQR
LF1 = q1-(1.5)*IQR
print(UF1)
print(LF1)
```

```python
Quantity=list(data['Quantity'])
Quantity_cols=[]
for i in range(len(Quantity)):
    if Quantity[i]<LF1:
        Quantity[i]=LF1
        Quantity_cols.append(Quantity[i])
    elif Quantity[i]>UF1:
        Quantity[i]=UF1
        Quantity_cols.append(Quantity[i])
    else:
        Quantity_cols.append(Quantity[i])
print(Quantity_cols)
```

```python
data['Quantity_cols']=Quantity_cols
```

```python
import seaborn as sns

sns.boxplot(Quantity_cols)
```
